```python
from torch.utils.data import DataLoader
from torchvision import transforms, datasets
import numpy as np
import pandas as pd
```

```python
# load CIFA-10 dataset
CIFAR_transform_train = transforms.Compose([transforms.ToTensor()])
CIFAR_transform_test =transforms.Compose([transforms.ToTensor()])
trainset_CIFAR = datasets.CIFAR10(root='./data', train=True, download=True, transform=
CIFAR_transform_train)
testset_CIFAR = datasets.CIFAR10(root='./data', train=False, download=True, transform=CIFAR_transform_test)
CIFAR_train = DataLoader(trainset_CIFAR, batch_size=32, shuffle=True, num_workers=2)
CIFAR_test = DataLoader(testset_CIFAR, batch_size=32, shuffle=False, num_workers=2)

CIFAR_train_images = []
CIFAR_train_labels = []
for batch in CIFAR_train:
    images, labels = batch
    images_flat = images.view(images.shape[0], -1)
    CIFAR_train_images.append(images_flat.numpy())
    CIFAR_train_labels.append(labels.numpy())

CIFAR_train_images = np.vstack(CIFAR_train_images)
CIFAR_train_labels = np.concatenate(CIFAR_train_labels)
CIFAR_test_images = []
CIFAR_test_labels = []

for batch in CIFAR_test:
    images, labels = batch
    images_flat = images.view(images.shape[0], -1)
    CIFAR_test_images.append(images_flat.numpy())
    CIFAR_test_labels.append(labels.numpy())
CIFAR_test_images = np.vstack(CIFAR_test_images)
CIFAR_test_labels = np.concatenate(CIFAR_test_labels)
```

```python
# load MNIST dataset
mnist_train_transform = transforms.Compose([transforms.ToTensor()])
mnist_test_transform = transforms.Compose([transforms.ToTensor()])
trainset_mnist = datasets.MNIST(root='./data', train=True, download=True, transform=mnist_train_transform)
testset_mnist = datasets.MNIST(root='./data', train=False, download=True, transform=mnist_test_transform)

MNIST_train = DataLoader(trainset_mnist, batch_size=32, shuffle=True, num_workers=2)
MNIST_test = DataLoader(testset_mnist, batch_size=32, shuffle=False, num_workers=2)

MNIST_train_images = []
MNIST_train_labels = []
for batch in MNIST_train:
    images, labels = batch
    images_flat = images.view(images.shape[0], -1)
    MNIST_train_images.append(images_flat.numpy())
    MNIST_train_labels.append(labels.numpy())

MNIST_train_images = np.vstack(MNIST_train_images)
MNIST_train_labels = np.concatenate(MNIST_train_labels)

MNIST_test_images = []
MNIST_test_labels = []
for batch in MNIST_test:
    images, labels = batch
    images_flat = images.view(images.shape[0], -1)
    MNIST_test_images.append(images_flat.numpy())
    MNIST_test_labels.append(labels.numpy())
MNIST_test_images = np.vstack(MNIST_test_images)
MNIST_test_labels = np.concatenate(MNIST_test_labels)
```

```python
# algorithms
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from scipy.stats import randint
from sklearn.model_selection import GridSearchCV

import pandas as pd

# config
tree_depth = [3,6,9,12]
params_grid = {
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_leaf_nodes': [5, 10, None]
}
```

```python
## MNIST

# four decision tree classifiers with each of depths
mnist_dt_classifiers = []
for i in tree_depth:
    gscv_dt = GridSearchCV(DecisionTreeClassifier(max_depth=i), params_grid, cv=5) # create a decision tree classifier
    mnist_dt_classifiers.append(gscv_dt)

# svm classifier
mnist_svm_lin_classifier = svm.SVC(kernel='linear')
mnist_svm_rbf_classifier = svm.SVC(kernel='rbf')
```

```python
## CIFAR10

# four decision tree classifiers with each of depths
cifar_dt_classifiers = []
for i in tree_depth:
    gscv_dt = GridSearchCV(DecisionTreeClassifier(max_depth=i), params_grid, cv=5) # create a decision tree classifier
    cifar_dt_classifiers.append(gscv_dt)

# svm classifier
cifar_svm_lin_classifier = svm.SVC(kernel='linear')
cifar_svm_rbf_classifier = svm.SVC(kernel='rbf')
```

```python
# dt fit
for i in range(len(mnist_dt_classifiers)):
    mnist_dt_classifiers[i].fit(MNIST_train_images, MNIST_train_labels)

# svm fit
mnist_svm_lin_classifier.fit(MNIST_train_images, MNIST_train_labels)
mnist_svm_rbf_classifier.fit(MNIST_train_images, MNIST_train_labels)
```

```python
mnist_dt_train_pred = []
mnist_dt_train_accuracy = []
mnist_dt_test_pred = []
mnist_dt_test_accuracy = []

mnist_svm_lin_train_pred = []
mnist_svm_lin_train_accuracy = []
mnist_svm_lin_test_pred = []
mnist_svm_lin_test_accyracy = []

mnist_svm_rbf_test_pred = []
mnist_svm_rbf_test_accuracy = []
mnist_svm_rbf_test_pred = []
mnist_svm_rbf_test_accuracy = []

# dt train
for i in range(len(mnist_dt_classifiers)):
    mnist_dt_train_pred.append(mnist_dt_classifiers[i].predict(MNIST_train_images))
    mnist_dt_train_accuracy.append(accuracy_score(MNIST_train_labels, mnist_dt_train_pred[i]))

# dt test
for i in range(len(mnist_dt_classifiers)):
    mnist_dt_test_pred.append(mnist_dt_classifiers[i].predict(MNIST_test_images))
    mnist_dt_test_accuracy.append(accuracy_score(MNIST_test_labels, mnist_dt_test_pred[i]))

# svm lin train
mnist_svm_lin_train_pred = mnist_svm_lin_classifier.predict(MNIST_train_images)
mnist_svm_lin_train_accuracy = accuracy_score(MNIST_train_labels, mnist_svm_lin_train_pred)

# svm lin test
mnist_svm_lin_test_pred = mnist_svm_lin_classifier.predict(MNIST_test_images)
mnist_svm_lin_test_accuracy = accuracy_score(MNIST_test_labels, mnist_svm_lin_test_pred)

# svm rbf train
mnist_svm_rbf_train_pred = mnist_svm_rbf_classifier.predict(MNIST_train_images)
mnist_svm_rbf_train_accuracy = accuracy_score(MNIST_train_labels, mnist_svm_rbf_train_pred)

# svm rbf test
mnist_svm_rbf_test_pred = mnist_svm_rbf_classifier.predict(MNIST_test_images)
mnist_svm_rbf_test_accuracy = accuracy_score(MNIST_test_labels, mnist_svm_rbf_test_pred)
```

```python
mnist_dt_df = pd.DataFrame(data={'tree_depth' : tree_depth, 'train_accuracy': mnist_dt_train_accuracy, 'test_accuracy': mnist_dt_test_accuracy})
mnist_dt_df
```

```python
kernel_type = ["linear", "rbf"]
svm_train_accuracy = [mnist_svm_lin_train_accuracy, mnist_svm_rbf_train_accuracy]
svm_test_accuracy = [mnist_svm_lin_test_accuracy, mnist_svm_rbf_test_accuracy]
mnist_svm_df = pd.DataFrame(data={'kernel_type' : kernel_type, 'train_accuracy' : svm_train_accuracy, 'test_accuracy' : svm_test_accuracy})
mnist_svm_df
```

```python
## CIFAR model fitting

# dt fit
for i in range(len(cifar_dt_classifiers)):
    cifar_dt_classifiers[i].fit(CIFAR_train_images, CIFAR_train_labels)

# svm fit
cifar_svm_lin_classifier.fit(CIFAR_train_images, CIFAR_train_labels)
cifar_svm_rbf_classifier.fit(CIFAR_train_images, CIFAR_train_labels)
```

```python
cifar_dt_train_pred = []
cifar_dt_train_accuracy = []
cifar_dt_test_pred = []
cifar_dt_test_accuracy = []

cifar_svm_lin_train_pred = []
cifar_svm_lin_train_accuracy = []
cifar_svm_lin_test_pred = []
cifar_svm_lin_test_accyracy = []

cifar_svm_rbf_test_pred = []
cifar_svm_rbf_test_accuracy = []
cifar_svm_rbf_test_pred = []
cifar_svm_rbf_test_accuracy = []

# dt train
for i in range(len(cifar_dt_classifiers)):
    cifar_dt_train_pred.append(cifar_dt_classifiers[i].predict(CIFAR_train_images))
    cifar_dt_train_accuracy.append(accuracy_score(CIFAR_train_labels, cifar_dt_train_pred[i]))

# dt test
for i in range(len(cifar_dt_classifiers)):
    cifar_dt_test_pred.append(cifar_dt_classifiers[i].predict(CIFAR_test_images))
    cifar_dt_test_accuracy.append(accuracy_score(CIFAR_test_labels, cifar_dt_test_pred[i]))

# svm lin train
cifar_svm_lin_train_pred = cifar_svm_lin_classifier.predict(CIFAR_train_images)
cifar_svm_lin_train_accuracy = accuracy_score(CIFAR_train_labels, cifar_svm_lin_train_pred)

# svm lin test
cifar_svm_lin_test_pred = cifar_svm_lin_classifier.predict(CIFAR_test_images)
cifar_svm_lin_test_accuracy = accuracy_score(CIFAR_test_labels, cifar_svm_lin_test_pred)

# svm rbf train
cifar_svm_rbf_train_pred = cifar_svm_rbf_classifier.predict(CIFAR_train_images)
cifar_svm_rbf_train_accuracy = accuracy_score(CIFAR_train_labels, cifar_svm_rbf_train_pred)

# svm rbf test
cifar_svm_rbf_test_pred = cifar_svm_rbf_classifier.predict(CIFAR_test_images)
cifar_svm_rbf_test_accuracy = accuracy_score(CIFAR_test_labels, cifar_svm_rbf_test_pred)
```
